# Redes Neuronales Recurrentes utilizadas para el análisis de acciones

## Comparación de arquitecturas optimizadas y no optimizadas para GPU

Nuestro objetivo de esta notebook es comparar 2 arquitecturas distintas de Redes Neuronales Recurrentes, la primera es la arquitectura "Long Short Time Memory" (LSTM) y la segunda "Gated Recurrent Unit" (GRU), esta segunda es una variación de la primera. Además, vamos a probar estas mismas arquitecturas aceleradas por CUDA, las arquitecturas aceleradas por CUDA solamente se pueden correr por GPU pero tiene una mejor performance que las arquitecturas no aceleradas sin perder precisión en la predicción. Cuando decimos aceleradas por CUDA nos referimos una API de computación paralela desarollada por nVidia para sos tarjetas graficas que permiten explotar las ventajas de las GPU frente a la CPU, esta plataforma es usado por cuDNN que es una libreria, también desarollada por nVidia, que hace uso de CUDA para accelerar el procesamiento de modelos de redes neuronales.

**Una aclaración teorica**: La predicción de las acciones por medio de estos métodos no siempre va a ser adecuada. Cuando se analizan acciones existe lo que se llama el análisis técnico y el análisis fundamental: 

**El análisis técnico**: Busca  predecir los movimientos de precios en los mercados financieros a partir de datos históricos y estadísticas de mercado.

**El análisis fundamental**: Es el analisis del contexto de las acciones, del valor intrínseco, esto incluye todos los aspectos posibles que no necesariamente se vean reflejados en datos historicos.

Cuando aplicamos nuestro modelo estamos obviando esta segunda parte, matematicamente se dice que las acciones tienen un movimiento browniano (Por lo menos parcialmente), es decir, el conocer toda la historia completa de las acciones no quiere decir que vayamos a poder predecir el futuro de estas.

*La explicación que se presenta de las distintas capas es muy breve y resumida, busca remarcar las diferencias entre ambas asumiendo que el lector conoce las arquitecturas. En el caso de no conocerlas recomiendo entrar en los links que deje para leer más sobre el tema*

**Long Short Time Memory**

Esta arquitectura es de las primeras que se comenzo a utilizar cuando se empezo a trabajar con redes neuronales recurrentes para contrarestar el efecto del gradiente desvaneciente. De forma resumida abajo vemos distintas capas, vemos que estan conectadas horizontalmente, pero los dato salen y entran verticalmente, los datos no viajan de la entrada directamente a la salida sino pasan por distintas capas, esto es lo que hace tan poderosas a las redes neuronales recurrentes los datos se alimentan de información anterior. Compensa la información que entra (nueva) con la información que proporciona la capa anterior. En https://colah.github.io/posts/2015-08-Understanding-LSTMs/ van a encontrar un mayor detalle del funcionamiento.

<img src="./LSTMarq.png">

**Gated Recurrent Unit**

Esta arquitectura es una variación de la anterior que la busca simplificar, vemos que ya el traspaso de información entre capas ahora es única, por lo general tanto la arquitectura LSTM como la GRU suelen tener resultados muy parecidos. 
A la hora de elegir uno o el otro se tiene en cuenta que la arquitectura GRU es más simple y suele performar mejor (Tarda menos tiempo), pero la arquitectura LSTM capta la memoria por separado a los resultados (Tiene dos puntos de traspaso de información) por lo tanto suele ser más adecuada cuando trabajamos con secuencias largas.
En https://medium.com/@george.drakos62/what-is-a-recurrent-nns-and-gated-recurrent-unit-grus-ea71d2a05a69 van a encontrar un mayor detalle del funcionamiento.

<img src="./GRUarq.png">

**Acerca del Dataset**: Este Dataset contiene la fecha del conjunto de datos, el valor de la acción a la apertura del mercado, al cierre, su valor máximo y su valor minimo de la jornada junto con el volumen comercializado. Para tener un modelo más preciso deberiamos de incluir más datos que no sean propias de la acción, como por ejemplo, el valor de acciones relacionadas o de indicadores externos. Para mantener el ejemplo simple nos quedamos con esta entrada.

El dataset de entrenamiento esta conformado con los datos mencionados historicos de las acciones de GOOGLE hasta el 2016-12-31, nuestro objetivo va a ser predecir el valor de las acciones para los proximos 20 días habiles. Para hacer esto vamos a entrenar nuestro modelo con todos estos datos teniendo en cuenta que hablamos de una red neuronal recurrente, es decir, vamos a brindar los datos por secuencias, para ejemplificar nuestro caso, siempre vamos a estar entregandole al modelo los datos del día 1 al 60 para predecir el valor de cierre del mercado del día 61.

Atención a que menciono días habiles, el mercado no abre todos los dias, por lo tanto van a tener en el dataset varias fechas que no se van a encontrar (Fin de semanas, feriados, etc.)

**Aclaración del equipo**: Por default Tensorflow, y por lo tanto Keras que corre por arriba de este, toman la GPU para el procesamiento. Este es el caso, por lo tanto, todos los modelos tanto los que usan capas aceleradas por CUDA como las que no corren por GPU.

## Fase 0: Importar librerías

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, CuDNNLSTM, CuDNNGRU, GRU
from tensorflow.python.client import device_lib
import keras.backend
import tensorflow as tf

import time

import gc

%matplotlib inline
```

## Fase 1: Entrenar modelo

### Cargar Dataset

```python
df_train = pd.read_csv('Google_Stock_Price_Train.csv', infer_datetime_format=True, decimal = '.', thousands=',')
df_train.columns.values.tolist()
```

### Cortamos la columna de fechas

```python
df_train = df_train.iloc[:,1:]
```

### Separamos en target y features

```python
df_train_x_col = df_train.drop(['Close'],axis=1).columns.values.tolist()

df_train_y = df_train['Close']
df_train_x = df_train.drop(['Close'],axis=1)

df_train_x.shape,df_train_y.shape
```

### Normalizamos target y features por separado

Normalizamos por separado ya que más adelante al hacer nuestra predicción necesitamos aplicar una normalización inversa a la predicción, sino nuestro resultado sería un conjunto de numeros normalizados que no podemos interpretar.

```python
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

df_train_norm_x = pd.DataFrame(scaler_x.fit_transform(df_train_x),
                              columns=df_train_x_col)
df_train_norm_y = pd.DataFrame(scaler_y.fit_transform(pd.DataFrame(df_train_y)),
                              columns=['Close'])

df_train_norm = pd.concat([df_train_norm_x,df_train_norm_y],axis=1)

df_train_norm.shape,df_train_norm.columns.values.tolist()
```

### Creamos el Dataset en el format requerido para la RNR

Cuando trabajamos con Redes Neuronales Recurrentes el dataset debe de tener un formato especial ya que nuestro input no son números sino vectores, estos vectores representan los ultimos n valores de cada feature (valor al cierre, volumen comercializado,...). Llevandolo a nuestro caso, nuestros inputs van a ser 5 features, cada uno es un vector que representa los ultimos 60 valores en orden cronologico.

```python
x_train = []
y_train = []
sequence_len = 60

for i in range(sequence_len,len(df_train_norm)):
    x_train.append(df_train_norm.values[i-sequence_len:i,:])
    y_train.append(df_train_norm.values[i,4])

x_train = np.asarray(x_train)
y_train = np.asarray(y_train)

y_train.shape,x_train.shape
```

### Verificamos que dispositivos encuentra Tensorflow

Lo que estamos viendo son los elementos de procesamiento que tenemos en la computadora y que tensorflow esta encontrando. En este caso tenemos la CPU y la GPU.

```python
print(device_lib.list_local_devices()) # list of DeviceAttributes
```

### Definimos la función de la Red Neuronal Recurrente

Este es el modelo de la Red Recurrente Neuronal, la arme como función para que la puedan copiar y hacer pruebas, tiene una breve explicación de los inputs. La función no carga las librerias, acurdense de copiarlas.

Las capas aceleradas por CUDA solamente funcionan cuando entrenamos el modelo por GPU, por lo tanto, si no tienen una GPU les va a dar un error, usen capas sin esta funcionalidad como LSTM o GRU.

```python
def _rnn_regressor_model(x, y, hidden_layer_amt=1, hidden_layer_type='LSTM', dropout=0.2, layer_units = [],\
                       output_units=1, optimizer='Adam', epochs_amt=100, batchsize=100):
    """
    Esta función nos permite generar una Red Neuronal Recurrente con algunos valores customizables.
    
    Input:
        x: Información para la predicción.
        y: Valor a predecir (Valor real).
        hidden_layer_amt: Cantidad de capas ocultas.
        hidden_layer_type: Tipo de Capa.
        dropout: Valor de dropout.
        layer_units: Vector con la cantidad de unidades por capa oculta, por default son 100 unidades por capa.
        output_units: Cantidad de elementos de output.
        optimizer: Optimizador.
        epochs_amt: Cantidad de iteraciones sobre el dataset.
        batchsize: Cantidad de filas de datos usados por iteración.
        
    Output:
        El modelo definido
        El modelo entrenado
    """
    ###Toma tiempo base###
    base_time = time.time()    
    
    ###Aplica 100 unidades para las capas con unidades no especificadas###
    for j in range(hidden_layer_amt-len(layer_units)):
        layer_units.append(100)
    
    ###Red Neuronal Recurrente###
    regressor_model = Sequential()

    regressor_model.add(hidden_layer_type(units = layer_units[0],\
                                          return_sequences=True,\
                                          input_shape = (x.shape[1],x.shape[2])))
    regressor_model.add(Dropout(dropout))

    for i in range(hidden_layer_amt-1):
        if (i == (hidden_layer_amt-2)):
            regressor_model.add(hidden_layer_type(units = layer_units[i+1], return_sequences=False))
            regressor_model.add(Dropout(dropout))
        else:
            regressor_model.add(hidden_layer_type(units = layer_units[i+1], return_sequences=True))
            regressor_model.add(Dropout(dropout))

    regressor_model.add(Dense(units = output_units))
    
    ###Compila la red y determina el optimizador y la función de perdida###
    regressor_model.compile(optimizer = optimizer, loss = 'mean_squared_error')
    
    ###Entrena el modelo###
    model_result = regressor_model.fit(x, y, epochs=epochs_amt, batch_size=batchsize,  verbose=0)
    
    ###Toma tiempo para el finalizado del proceso###
    time_1 = time.time()
    
    ###Info del procesamiento (Tiempo y verificación si corrio con CPU o GPU)###
    print('El modelo tardo {:.2f} minutos para entrenarse.'.format((time_1-base_time)/60))
    
    if len(keras.backend.tensorflow_backend._get_available_gpus()) == 0:
        print('Se utilizo la CPU para entrenar este modelo')
    else:
        print('Se utilizo la GPU para entrenar este modelo')
    
    return regressor_model, model_result
```

### Entrenamos la red

Comparamos cuanto tarda en entrenar el modelo usando LSTM, GRU, CuDNNLSTM y CuDNNGRU

Cuando usamos la GPU es importante mirar el batch size, una de las ventajas de la GPU frente al CPU es que el primero tiene una gran cantidad de núcleos, no así la CPU, pero los núcleos de la CPU son mucho más potentes a los núcleos de la GPU.
Por lo tanto, si trabajamos con un batch size muy chico, es decir, que corren pocos procesos en simultaneo probablemente es más recomendable correr el modelo por CPU, ahora si el batch size es grande la performance de la GPU siempre va a ser mejor. Esto no quiere decir que debemos de entrenar el modelo con todos los datos juntos, si estudiaron la teoría de redes neuronales saben que el usar un batch para entrenar el modelo es un proceso estocastico que nos sirve para no optimizar a minimos locales de la función de perdida, sino llegar al minimo global.

#### Usando capa "Long Short Time Memory"

```python
regressor_model_LSTM, model_result_LSTM = _rnn_regressor_model(x_train, y_train, 5, LSTM, batchsize=100)
```

#### Usando capa "Gated Recurrent Unit"

```python
regressor_model_GRU, model_result_GRU = _rnn_regressor_model(x_train, y_train, 5, GRU, batchsize=100)
```

#### Usando capa "Long Short Time Memory" acelerado via CUDA

```python
regressor_model_CuDNNLSTM, model_result_CuDNNLSTM = _rnn_regressor_model(x_train, y_train, 5, CuDNNLSTM, batchsize=100)
```

#### Usando capa "Gated Recurrent Unit" acelerado via CUDA

```python
regressor_model_CuDNNGRU, model_result_CuDNNGRU = _rnn_regressor_model(x_train, y_train, 5, CuDNNGRU, batchsize=100)
```

## Fase 2: Validamos el modelo

### Cargmos los datasets para la validación

```python
df_historic = pd.read_csv('Google_Stock_Price_Train.csv',infer_datetime_format=True,decimal = '.', thousands=',')
df_2017 = pd.read_csv('Google_Stock_Price_Test.csv',infer_datetime_format=True,decimal = '.', thousands=',')
```

### Formateamos el dataset para obtener los datos necesarios

```python
df = pd.concat([df_historic,df_2017],axis=0)
df_complete = df.iloc[:,1:]

df_test = df_complete[len(df_historic)-60:]

df_test.shape
```

### Separamos en target y features

```python
df_test_x_col = df_test.drop(['Close'],axis=1).columns.values.tolist()

df_test_y = df_test['Close']
df_test_x = df_test.drop(['Close'],axis=1)

df_test_x.shape,df_test_y.shape
```

### Normalizamos target y features por separado usando el escalador anterior

Noten que no estoy usando la función .fit_transform, sino .transform esto es porque estoy normalizando con los mismos parametros que use para el dataset de entrenamiento.

```python
df_test_norm_x = pd.DataFrame(scaler_x.transform(df_test_x),
                              columns=df_test_x_col)
df_test_norm_y = pd.DataFrame(scaler_y.transform(pd.DataFrame(df_test_y)),
                              columns=['Close'])

df_test_norm = pd.concat([df_test_norm_x,df_test_norm_y],axis=1)

df_test_norm.shape,df_test_norm.columns.values.tolist()
```

### Creamos el Dataset en el formato usado para la RNR

```python
x_test = []

for i in range(sequence_len,len(df_test_norm)):
    x_test.append(df_test_norm.values[i-sequence_len:i,:])

x_test = np.asarray(x_test)

x_test.shape
```

### Predicción y desnormalización

```python
pred_close_price_LSTM = regressor_model_LSTM.predict(x_test)
pred_close_price_LSTM = scaler_y.inverse_transform(pred_close_price_LSTM)
```

```python
pred_close_price_GRU = regressor_model_GRU.predict(x_test)
pred_close_price_GRU = scaler_y.inverse_transform(pred_close_price_GRU)
```

```python
pred_close_price_CuDNNLSTM = regressor_model_CuDNNLSTM.predict(x_test)
pred_close_price_CuDNNLSTM = scaler_y.inverse_transform(pred_close_price_CuDNNLSTM)
```

```python
pred_close_price_CuDNNGRU = regressor_model_CuDNNGRU.predict(x_test)
pred_close_price_CuDNNGRU = scaler_y.inverse_transform(pred_close_price_CuDNNGRU)
```

## Fase 3: Analísis

### Graficamos las predicciones y los valores reales

En el grafico debemos de observar la tendencia de nuestras curvas de predicción, por lo que mencionamos al principio la predicción puede ser más a menos precisa dependiendo de cuanta información se nos esta escapando, pero si es importante que nuestras curvas predigan bien la tendencia. Cuando uno invierte debe de comprar acciones y venderlas a un precio mayor basicamente, es decir, debemos de saber si la acción mañana va a tener una tendencia de baja para no comprar hoy y esperar,  o si va a estar en alza y debemos de comprar ya. Si vemos las curvas parece que las dos curvas armadas con una arquitectura GRU captan de forma más sensible la tendencia, tiene sentido ya que la información historica tiene menos peso en esta arquitectura.

### Calculamos el RMSE y el RMSE relativo (Root mean squared error)

Esta métrica nos permite ver que tan cerca estuvo nuestra predicción del valor real. Vemos el error como un valor absoluto y el relativo, esto es ese mismo valor dividido por el promedio del valor real de la acción durante los primeros 20 días del 2017, los días para los cuales realizamos la predicción. Como estamos analizando datos secuenciales este valor relativo no es suficiente para determinar que tan bueno es nuestro modelo, deberiamos de ver que diferencia relativa tuvimos durante el mes pero a nivel diario, de todas formas si nos sirve para darnos una idea general.

```python
rmse_LSTM = np.sqrt(sum((pred_close_price_LSTM-df_2017['Close'].values.reshape(-1,1))**2)).squeeze()
rmse_GRU = np.sqrt(sum((pred_close_price_GRU-df_2017['Close'].values.reshape(-1,1))**2)).squeeze()
rmse_CuDNNLSTM = np.sqrt(sum((pred_close_price_CuDNNLSTM-df_2017['Close'].values.reshape(-1,1))**2)).squeeze()
rmse_CuDNNGRU = np.sqrt(sum((pred_close_price_CuDNNGRU-df_2017['Close'].values.reshape(-1,1))**2)).squeeze()

relative_LSTM = (np.sqrt(sum((pred_close_price_LSTM-df_2017['Close'].values.reshape(-1,1))**2))/np.mean(df_2017['Close'])).squeeze()
relative_GRU = (np.sqrt(sum((pred_close_price_GRU-df_2017['Close'].values.reshape(-1,1))**2))/np.mean(df_2017['Close'])).squeeze()
relative_CuDNNLSTM = (np.sqrt(sum((pred_close_price_CuDNNLSTM-df_2017['Close'].values.reshape(-1,1))**2))/np.mean(df_2017['Close'])).squeeze()
relative_CuDNNGRU = ((np.sqrt(sum((pred_close_price_CuDNNGRU-df_2017['Close'].values.reshape(-1,1))**2))/np.mean(df_2017['Close'])).squeeze())

print('RMSE para el modelo con capas LSTM \t\t{:.2f}\t El error relativo es de {:.2f}%'.format(rmse_LSTM,relative_LSTM))
print('RMSE para el modelo con capas GRU \t\t{:.2f}\t El error relativo es de {:.2f}%'.format(rmse_GRU,relative_GRU))
print('RMSE para el modelo con capas CuDNNLSTM \t{:.2f}\t El error relativo es de {:.2f}%'.format(rmse_CuDNNLSTM,relative_CuDNNLSTM))
print('RMSE para el modelo con capas CuDNNGRU \t\t{:.2f}\t El error relativo es de {:.2f}%'.format(rmse_CuDNNGRU,relative_CuDNNGRU))
```

### Calculamos el error relativo por dia

En el grafico vimos que tenemos un valor fuera de tendencia seguida de una caida en el valor real de las acciones del GOOGLE, estos valores fuera de rango son dificiles de predecir por un modelo de este tipo, esto es lo que vemos en el error relativo por día, el mayor error, en todos los modelos, se produce cuando tenemos este pico seguido por la caida.

```python
daily_relative_LSTM = ((df_2017['Close'].values.reshape(-1,1)-pred_close_price_LSTM)/df_2017['Close'].values.reshape(-1,1)).squeeze()
daily_relative_GRU = ((df_2017['Close'].values.reshape(-1,1)-pred_close_price_GRU)/df_2017['Close'].values.reshape(-1,1)).squeeze()
daily_relative_CuDNNLSTM = ((df_2017['Close'].values.reshape(-1,1)-pred_close_price_CuDNNLSTM)/df_2017['Close'].values.reshape(-1,1)).squeeze()
daily_relative_CuDNNGRU = ((df_2017['Close'].values.reshape(-1,1)-pred_close_price_CuDNNGRU)/df_2017['Close'].values.reshape(-1,1)).squeeze()

plt.figure(figsize=(30,16))
plt.title('Error relativo diario',fontdict={'weight':'bold','size':40})
plt.xlabel('Fecha',fontdict={'size':30,'weight':'bold'})
plt.ylabel('Error relativo (%)',fontdict={'size':30,'weight':'bold'})

plt.plot(df_2017['Date'],daily_relative_LSTM,c='orange',ls='--',alpha=0.5,label='Predicción LSTM',lw=3)
plt.plot(df_2017['Date'],daily_relative_GRU,c='c',ls='--',alpha=0.5,label='Predicción GRU',lw=3)
plt.plot(df_2017['Date'],daily_relative_CuDNNLSTM,c='orange',alpha=0.5,label='Predicción CuDNNLSTM',lw=3)
plt.plot(df_2017['Date'],daily_relative_CuDNNGRU,c='c',alpha=0.5,label='Predicción CuDNNGRU',lw=3)
plt.hlines(y=0,xmin=df_2017.iloc[0,0],xmax=df_2017.iloc[-1,0],colors='r',lw=5, alpha=0.5)
plt.legend(prop={'size':30})
ax = plt.gca()
ax.set_facecolor('k')
plt.grid(color='#666666')
plt.xticks(size=20, rotation=60)
plt.yticks(size=20)
plt.show()
```

Las capas aceleradas son claramente superiores respecto al tiempo de procesamiento y los resultados no parecen verse afectados por el uso de estas capas. Si vemos que las capas GRU son ligeramente más precisas. Podemos decir que se cumple todo aquello que se supuso de antemano.

## Fase 4: Selección del mejor modelo (Extra)

En esta parte les voy a dar un ejemplo de como iterar distintos valores sobre un hiperparametro.

Por el tiempo de procesamiento y el resultado, el modelo que mejor nos resulto fue el que utilizo la arquitectura CuDNNGRU, ahora vamos a usar este modelo para ajustar otros hiperparametros en el. Hay una función usada para esto que se llama GridSearchCV esta pertenece a la librería de scikit-learn, esta libreria no tiene implementada el uso de GPU, por lo tanto vamos a realizar una comprobación parecida obviando la cross validation de forma manual.

### Agregamos librerias necesarias

```python
from sklearn.model_selection import ParameterGrid
```

### Definimos los parametros a probar

En formato de diccionario estoy indicando que hiperparametro de la función de mi red neuronal recurrente voy a iterar y que valores quiero probar.

```python
to_opt_params = {
    'hidden_layer_amt': [2,6,12],
}

print('Vamos a comparar {} modelos'.format(len(list(ParameterGrid(to_opt_params)))))
```

### Buscamos optimizar los hiperparametros del modelo

En este ejemplo voy a probar distintas profundidades de red usando 2, 6 o 12 capas ocultas para analizar con cual obtenemos un menor valor RMSE.

```python
results_vec=[]

time_1 = time.time()

for i in range(len(list(ParameterGrid(to_opt_params)))):
    list(ParameterGrid(to_opt_params))[i]['hidden_layer_amt']
    
    time_2 = time.time()
    
    print('Comienza el entrenamiento del modelo {}\n'.format(list(ParameterGrid(to_opt_params))[i]))
    
    model, result = _rnn_regressor_model(x_train, y_train,\
                        hidden_layer_type=CuDNNGRU,\
                        hidden_layer_amt=list(ParameterGrid(to_opt_params))[i]['hidden_layer_amt'])
    time_3 = time.time()
    
    pred = model.predict(x_test)
    pred = scaler_y.inverse_transform(pred)
    rmse = np.sqrt(sum((pred-df_2017['Close'].values.reshape(-1,1))**2)).squeeze()
    
    
    print('RMSE: {:.4f}'.format(rmse))
    print('Faltan {} modelos más\n'.format(len(list(ParameterGrid(to_opt_params)))-i-1))
    
    results_vec.append([model,result,rmse])
        
time_4 = time.time()
print('El proceso completo tomo {:.2f} minutos'.format((time_4-time_1)/60))
```

Obtuvimos el mejor resultado usando la menor cantidad de capas ocultas (A veces menos es más).

### Graficamos el resultado del modelo final
